```python
import numpy as np
import matplotlib.pyplot as plt

# Define Moseley's Law constant (a) and element atomic numbers
a = 2.47e15  # Proportionality constant in Hz^(1/2), approximation
elements = {
    "Copper": 29,
    "Rubidium": 37,
    "Molybdenum": 42,
    "Silver": 47,
    "Barium": 56,
    "Terbium": 65
}

# Screening constants for K_alpha and K_beta transitions
b_values = {
    "K_alpha": 1.0,  # Screening constant for n=2 -> n=1
    "K_beta": 7.4    # Screening constant for n=3 -> n=1 (approximation)
}

# Function to calculate frequencies based on Moseley's law
def calculate_frequencies(Z, a, b):
    sqrt_nu = a * (Z - b)
    return sqrt_nu**2

# Calculate frequencies for K_alpha and K_beta lines
frequencies = {"K_alpha": {}, "K_beta": {}}
for line, b in b_values.items():
    for element, Z in elements.items():
        freq = calculate_frequencies(Z, a, b)
        frequencies[line][element] = freq

# Display results
print("X-ray frequencies for selected elements (in Hz):")
for line, data in frequencies.items():
    print(f"\n{line} lines:")
    for element, freq in data.items():
        print(f"  {element}: {freq:.2e} Hz")

# Plot results
plt.figure(figsize=(12, 8))

# Plot for K_alpha and K_beta
for line, b in b_values.items():
    atomic_numbers = list(elements.values())
    frequency_values = [frequencies[line][el] for el in elements]
    plt.plot(atomic_numbers, frequency_values, marker='o', label=f"{line} Line")

# Plot settings
plt.title("Moseley's Law for K_alpha and K_beta Lines")
plt.xlabel("Atomic Number (Z)")
plt.ylabel("Frequency (Hz)")
plt.grid(True)
plt.legend()
plt.show()

```
